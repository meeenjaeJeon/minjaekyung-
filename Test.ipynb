```python
import tensorflow as tf
import numpy as np
from functools import partial

saver = tf.train.import_meta_graph("./my_model_final.ckpt.meta")

x = np.loadtxt("TEST_data.csv", delimiter=',', dtype=np.float32)
y = np.loadtxt("test_solution.csv", delimiter=',', dtype=np.float32).reshape(418,1)

with tf.Session() as sess:
    saver.restore(sess, "./my_model_final.ckpt")
    X = tf.get_default_graph().get_tensor_by_name("X:0")
    Y = tf.get_default_graph().get_tensor_by_name("Y:0")
    predicted = tf.get_default_graph().get_tensor_by_name("eval/Sigmoid:0")
    accuracy = tf.get_default_graph().get_tensor_by_name("eval/Mean:0")
    training_op = tf.get_default_graph().get_operation_by_name("train/RMSProp")
    pre = predicted.eval(feed_dict={X:x})
    acc = accuracy.eval(feed_dict={X:x, Y:y})
    print("acc:", acc)
```

```python
for op in tf.get_default_graph().get_operations():
    print(op.name)
```

```python
sub = np.round(pre).astype(int).reshape(418,1)
sub = sub.flatten()
```

```python
import pandas as pd
df1 = pd.DataFrame({"PassengerId": np.arange(892,1310), "Survived": sub})
df2 = df1.set_index("PassengerId")
df2
```
